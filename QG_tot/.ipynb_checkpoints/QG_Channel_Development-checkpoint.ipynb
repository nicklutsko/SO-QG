```python
import numpy as np
import random
random.seed(100)
import matplotlib.pyplot as plt
import math
from QG.io import NetCDFWriter
import scipy.fft as sci_fft
fft = sci_fft
nworker=10
```

```python
class QG_Channel(object):
    """
    Dynamical core for a spectral QG model in a channel.
    equation model.
    """
    
    def __init__(self, N, N2, dt,opt = 3, seed=1, 
                 robert_coefficient=0.04):
        """
        Initialize a barotropic model.
        Arguments:
        * vrt : numpy.ndarray[nlat, nlon]
            An initial vorticity field on the desired model grid. The
            grid is assumed to be Gaussian grid, and in general nlon is
            double nlat.
        * truncation : int
            The spectral truncation (triangular). A suggested value is
            nlon // 3.
        * dt : float
            The model time-step in seconds.
        * start_time : datetime.datetime
            A datetime object representing the start time of the model
            run. This doesn't affect computation, it is only used for
            metadata.
        Optional arguments:
        * robert_coefficient : default 0.04
            The coefficient for the Robert time filter.
        """
        # Model grid size:
        self.N=N
        self.N2=N2
        #we should instantiate with Lx and Ly I think
        self.Lx = 72.
        self.Ly = 96.
        self.x = np.linspace( -self.Lx / 2, self.Lx / 2, N )
        self.y = np.linspace( -self.Ly / 2, self.Ly / 2, N2 ) #domain size
        self.ll = np.fft.fftfreq( self.N2, self.Ly / float(self.N2) / 2. / np.pi ) #meridional wavenumbers
        self.kk = np.fft.rfftfreq( self.N, self.Lx / float(self.N) / 2. / np.pi ) #zonal wavenumbers
        self.dt = dt #time step
        
        self.nu = pow(10.,-3.) #viscous dissipation
        self.tau_d = 100. #Newtonian relacation time scale for interface 
        self.tau_f = 15. #surface friction
        self.beta = 0.196 #beta in JH
        self.sigma = 3.5
        self.U_1=1.
        self.robert_coefficient = robert_coefficient
        self.dx= min([ np.diff(self.x)[0], np.diff(self.y)[0] ])
        self.Lapl = ( self.kk[np.newaxis, :]** 2  + self.ll[:, np.newaxis]** 2 )
        
        #linearity condition
        self.opt=3
        
        #time options
        self.first_step = True
        self.t=0
        self.count=0
        self.CFL = 0 
        
        
        self.km = 1. / 5.4 #surface friction, corresponds to 25 days in JH
        self.U = 1. #Maximum upper level wind speed
        #Running EKE spectrum
        self.ta = 100 #How often to calculate spectra
        self.lim = 1 #when to start calculating spectra

        #Spectral arrays, only need 3 time-steps
        self.psic_1 = np.zeros( ( ( 3, N2, N // 2 + 1 ) ) ).astype( complex )
        self.psic_2 = np.zeros( ( ( 3, N2, N // 2 + 1  ) ) ).astype( complex )
        self.qc_1 = np.zeros( ( ( 3, N2, N // 2 + 1  ) ) ).astype( complex )
        self.qc_2 = np.zeros( ( ( 3, N2, N // 2 + 1  ) ) ).astype( complex )
        self.vorc_1 = np.zeros( ( ( 3, N2, N // 2 + 1  ) ) ).astype( complex )
        self.vorc_2 = np.zeros( ( ( 3, N2, N // 2 + 1  ) ) ).astype( complex )
        print('complex array shapes:',self.psic_2.shape)
        
        #real arrays
        #Real arrays, only need 3 time-steps
        self.psi_1 = np.zeros( ( ( 3 , N2 , N ) ) )
        self.psi_2 = np.zeros( ( ( 3 , N2 , N ) ) )
        self.q_1 = np.zeros( ( ( 3 , N2, N ) ) )
        self.q_2 = np.zeros( ( ( 3 , N2 , N ) ) )
        print('real array shapes:',self.q_1.shape)

        
        
        #think about making this a function so you can change it w/o initializing a new model.
        self.sponge= np.zeros(N2)
        self.u_eq = np.zeros(N2)
        for i in range( N2 ):
            y1 = float( i - N2 /2) * (self.y[1] - self.y[0] )
            y2 = float(min(i, N2 -i - 1)) * (self.y[1] - self.y[0] )
            self.sponge[i] = self.U_1 / (np.cosh(abs(y2/self.sigma)))**2
            self.u_eq[i] = self.U_1 * ( 1. / (np.cosh(abs(y1/self.sigma)))**2 - 1. / (np.cosh(abs(y2/self.sigma)))**2)
        
        
        self.psi_Rc = -fft.fft(  self.u_eq ) / 1.j / self.ll
        self.psi_Rc[0] = 0.
        self.psi_R = fft.ifft(self.psi_Rc)
        
        self.L = 0 #Linearized Clausius Claperon parameter
        self.C = 0 # non-dimensional measure of the strength of latent heating 
          
            
            
    def ptq(self):
        """Calculate PV"""
        self.qc_1[0] = -(self.ll[:, np.newaxis] ** 2 + self.kk[np.newaxis, :] ** 2 ) * self.psic_1[0] - (self.psic_1[0] - self.psic_2[0]) / 2. # -(k^2 + l^2) * psi_1 -0.5*(psi_1-psi_2)
        self.qc_2[0] = -(self.ll[:, np.newaxis] ** 2 + self.kk[np.newaxis, :] ** 2 ) * self.psic_2[0] + (self.psic_1[0] - self.psic_2[0]) / 2. # -(k^2 + l^2) * psi_2 +0.5*(psi_1-psi_2)
        self.qc_1[1] = -(self.ll[:, np.newaxis] ** 2 + self.kk[np.newaxis, :] ** 2 ) * self.psic_1[1] - (self.psic_1[1] - self.psic_2[1]) / 2. # -(k^2 + l^2) * psi_1 -0.5*(psi_1-psi_2)
        self.qc_2[1] = -(self.ll[:, np.newaxis] ** 2 + self.kk[np.newaxis, :] ** 2 ) * self.psic_2[1] + (self.psic_1[1] - self.psic_2[1]) / 2. # -(k^2 + l^2) * psi_2 +0.5*(psi_1-psi_2)
        self.vorc_1[0] = -(self.ll[:, np.newaxis] ** 2 + self.kk[np.newaxis, :] ** 2 ) * self.psic_1[0] - (self.psic_1[0] - self.psic_2[0]) / 2. # -(k^2 + l^2) * psi_1 -0.5*(psi_1-psi_2)
        self.vorc_2[0] = -(self.ll[:, np.newaxis] ** 2 + self.kk[np.newaxis, :] ** 2 ) * self.psic_2[0] + (self.psic_1[0] - self.psic_2[0]) / 2. # -(k^2 + l^2) * psi_2 +0.5*(psi_1-psi_2)
        self.vorc_1[1] = -(self.ll[:, np.newaxis] ** 2 + self.kk[np.newaxis, :] ** 2 ) * self.psic_1[1] - (self.psic_1[1] - self.psic_2[1]) / 2. # -(k^2 + l^2) * psi_1 -0.5*(psi_1-psi_2)
        self.vorc_2[1] = -(self.ll[:, np.newaxis] ** 2 + self.kk[np.newaxis, :] ** 2 ) * self.psic_2[1] + (self.psic_1[1] - self.psic_2[1]) / 2. # -(k^2 + l^2) * psi_2 +0.5*(psi_1-psi_2)
        
        
    def qtp(self):
        """Invert PV CHECK THIS"""
        psi_bt = -(self.vorc_1[2] + self.vorc_2[2]) / (self.ll[:, np.newaxis] ** 2 + self.kk[np.newaxis, :] ** 2) / 2.  # (psi_1 + psi_2)/2
        psi_bc = -(self.vorc_1[2] - self.vorc_2[2]) / (self.ll[:, np.newaxis] ** 2 + self.kk[np.newaxis, :] ** 2 + 1. ) / 2.  # (psi_1 - psi_2)/2
        psi_bt[0, 0] = 0.
        self.psic_1[1] = psi_bt + psi_bc
        self.psic_2[1] = psi_bt - psi_bc
            
        psi_bt = -(self.vorc_1[1] + self.vorc_2[1]) / (self.ll[:, np.newaxis] ** 2 + self.kk[np.newaxis, :] ** 2) / 2.  # (psi_1 + psi_2)/2
        psi_bc = -(self.vorc_1[1] - self.vorc_2[1]) / (self.ll[:, np.newaxis] ** 2 + self.kk[np.newaxis, :] ** 2 + 1. ) / 2.  # (psi_1 - psi_2)/2
        psi_bt[0, 0] = 0.
        self.psic_1[0] = psi_bt + psi_bc
        self.psic_2[0] = psi_bt - psi_bc
            
            
    def qtpM(self):
        """Invert PV"""
        divider =  ( np.expand_dims(self.ll, 1) ** 2 + np.expand_dims(self.kk, 0) ** 2)  # (psi_1 + psi_2)/2
        divider[0, 0] = np.nan
        psi_bt = -(self.vorc_1[1] + self.vorc_2[1]) / divider / 2.0  # (psi_1 + psi_2)/2
        psi_bt[0, 0] = 0.

        psi_bc = -(self.vorc_1[1] - self.vorc_2[1]) / (np.expand_dims(self.ll, 1) ** 2 + np.expand_dims(self.kk, 0) ** 2 + 2. ) / 2.  # (psi_1 - psi_2)/2
        self.psic_1[1] = psi_bt + psi_bc
        self.psic_2[1] = psi_bt - psi_bc
        
        
    def qtpM_3d(self):
        """Invert PV"""
        divider =  ( np.expand_dims(self.ll, 1) ** 2 + np.expand_dims(self.kk, 0) ** 2)  # (psi_1 + psi_2)/2
        divider[0, 0] = np.nan
        psi_bt = -(self.vorc_1[0:2] + self.vorc_2[0:2]) / divider / 2.0  # (psi_1 + psi_2)/2
        psi_bt[:,0, 0] = 0.

        psi_bc = -(self.vorc_1[0:2] - self.vorc_2[0:2]) / (np.expand_dims(self.ll, 1) ** 2 + np.expand_dims(self.kk, 0) ** 2 + 2. ) / 2.  # (psi_1 - psi_2)/2
        self.psic_1[0:2] = psi_bt + psi_bc
        self.psic_2[0:2] = psi_bt - psi_bc
    
        
    def random_init(self):
        
        print('starting from scratch')
        self.psic_1[0] = [ [ random.random() for i in range(self.N // 2 + 1) ] for j in range(self.N2) ]
        self.psic_2[0] = [ [ random.random() for i in range(self.N // 2 + 1) ] for j in range(self.N2) ]
        
        #Transfer values:
        self.psic_1[ 1 , : , : ] = self.psic_1[ 0 , : , : ]
        self.psic_2[ 1 , : , : ] = self.psic_2[ 0 , : , : ]
        
        
        #Calculate initial PV
        self.ptq()
        self.q_1[0:2, :,: ] = fft.irfft2( self.vorc_1[0:2,:,:], workers=nworker ) + self.beta * self.y[:, np.newaxis] # to real
        self.q_2[0:2, :,: ] = fft.irfft2( self.vorc_2[0:2,:,:], workers=nworker ) + self.beta * self.y[:, np.newaxis] # to real
        self.qc_1[0:2,:,:] = fft.rfft2( self.q_1[0:2,:,:] , workers=nworker )  # to imag
        self.qc_2[0:2,:,:] = fft.rfft2( self.q_2[0:2,:,:] , workers=nworker )  # to imag
        
        self.psi1 = fft.irfft2(self.psic_1[1] , workers=nworker )
        self.psi2 = fft.irfft2(self.psic_2[1] , workers=nworker )
        
        
        
    
    def integrate(self):
        """Step the model forward in time by one time-step."""
        
        #calcrhs(qc1,qc2,q_1,q_2,vorc_1,vorc_2,psi1,psi2,psic_1,psic_2,psi_R,N,N2,ll,kk,tau_d,tau_f,sponge,beta,opt)
        
        [rhs1,rhs2,cfor1,cfor2] = calcrhs(self.qc_1,self.qc_2,self.q_1,self.q_2,self.vorc_1,self.vorc_2,
                              self.psi_1,self.psi_2,self.psic_1, self.psic_2, self.psi_R,
                              self.N,self.N2,self.ll,self.kk,self.tau_d,self.tau_f,self.sponge,self.beta,self.opt)
        if self.first_step:
            #Forward step
            self.qc_1[2, :] = fs(self.qc_1[1, :, :], rhs1[:],self.dt,self.kk,self.ll,self.nu)
            self.qc_2[2, :] = fs(self.qc_2[1, :, :], rhs2[:],self.dt,self.kk,self.ll,self.nu)
            self.first_step = False
        
        else:
            self.qc_1[2, :, :] = lf(self.qc_1[0, :, :], rhs1[:], self.dt,self.ll,self.kk,self.nu)
            self.qc_2[2, :, :] = lf(self.qc_2[0, :, :], rhs2[:], self.dt,self.ll,self.kk,self.nu)
            

        self.q_1[1] = fft.irfft2( self.qc_1[2]  , workers =nworker)
        self.q_2[1] = fft.irfft2( self.qc_2[2]  , workers =nworker)
        
        #Subtract off BETA and Invert:
        self.vorc_1[1] = fft.rfft2( self.q_1[1] - self.beta * self.y[:, np.newaxis] , workers =nworker)
        self.vorc_2[1] = fft.rfft2( self.q_2[1] - self.beta * self.y[:, np.newaxis] , workers =nworker)
        self.qtpM() #psic_1[1], psic_2[1] = qtp(kk,ll, vorc_1[1], vorc_2[1] )
        self.psi_1[1] = fft.irfft2( self.psic_1[1]  , workers =nworker)
        self.psi_2[1] = fft.irfft2( self.psic_2[1]  , workers =nworker)
        
        ###########INSERT MOIST CALCS HERE: #############
        
        #################################################
        
        self.q_1[0:2] = fft.irfft2( self.qc_1[1:]  , workers =nworker)
        self.q_2[0:2] = fft.irfft2( self.qc_2[1:]  , workers =nworker)
        
        #Subtract off beta and invert
        self.vorc_1[0:2] = fft.rfft2( self.q_1[0:2] - self.beta * self.y[:, np.newaxis] , workers =nworker)
        self.vorc_2[0:2] = fft.rfft2( self.q_2[0:2] - self.beta * self.y[:, np.newaxis] , workers =nworker)
        
        self.qtpM_3d() #self.psic_1[0:2], psic_2[0:2] = qtp_3d(kk, ll, vorc_1[0:2], vorc_2[0:2] )
        
        
        #leap frog filter.
        if not self.first_step:
            self.qc_1[1, :] = filt( self.qc_1[1, :], self.qc_1[0, :], self.qc_1[2, :],self.robert_coefficient)
            self.qc_2[1, :] = filt( self.qc_2[1, :], self.qc_2[0, :], self.qc_2[2, :],self.robert_coefficient)
        
#         self.qtp()
        #Transfer values:
        self.qc_1[0:2, :, :] = self.qc_1[1:, :, :]
        self.qc_2[0:2, :, :] = self.qc_2[1:, :, :]
        self.t += self.dt
        self.count += 1
        

##### needed functions ###########
#######################################################
#  Spectral functions
#######################################################

#  Time-stepping functions
def reshuffle_helper1(psi, qc):
    N2, N = np.shape( psi )
    ex = int(N *  3 / 2)# - 1
    ex2 = int(N2 * 3 / 2)# - 1
    temp1 = np.zeros( ( ex2, ex ) ) + 0.j
    temp2 = np.zeros( ( ex2, ex ) ) + 0.j
    temp4 = np.zeros( ( N2, N ) ) + 0.j	#Final array

    #Pad values:
    temp1[:N2//2, :N] = psi[:N2//2, :N]
    temp1[ex2-N2//2:, :N] = psi[N2//2:, :N]

    temp2[:N2//2, :N] = qc[:N2//2, :N]
    temp2[ex2-N2//2:, :N] = qc[N2//2:, :N]
    return temp1, temp2, temp4, N2, N, ex2

def reshuffle_helper2(temp3, temp4, N2, N, ex2 ):
    temp4[:N2//2, :N] = temp3[:N2//2, :N]
    temp4[N2//2:, :N] = temp3[ex2-N2//2:, :N]
    return temp4

def calc_nl( psi, qc ):
    """"Calculate non-linear terms, with Orszag 3/2 de-aliasing"""

    temp1, temp2, temp4 , N2, N, ex2 = reshuffle_helper1(psi, qc)
    #Fourier transform product, normalize, and filter:
    temp3 = fft.rfft2( fft.irfft2( temp1 , workers=nworker ) * fft.irfft2( temp2, workers=nworker ) , workers=nworker ) * 9. / 4.

    return reshuffle_helper2(temp3, temp4, N2, N, ex2)

def jacobian_prep(kk, ll, psi, qc):
    #kk, ll, psi, qc = kk, ll, psic_1[1, :, :], vorc_1[1, :, :]

    kk2, ll2 =np.expand_dims(kk, 0), np.expand_dims(ll, 1)
    dpsi_dx = 1.j * kk2 * psi
    dpsi_dy = 1.j * ll2 * psi

    dq_dx = 1.j * kk2 * qc
    dq_dy = 1.j * ll2 * qc
    return dpsi_dx, dq_dy, dpsi_dy ,dq_dx

def nlterm(kk, ll, psi, qc):
    """"Calculate Jacobian"""

    dpsi_dx, dq_dy, dpsi_dy ,dq_dx = jacobian_prep(kk, ll, psi, qc)  ######### here was an error

    return  calc_nl( dpsi_dx, dq_dy ) - calc_nl( dpsi_dy, dq_dx )

def fs(ovar, rhs, det,kk,ll,nu):
    """Forward Step: q^t-1 / ( 1 + 2. dt * nu * (k^4 + l^4 ) ) + RHS"""
    mult = det / ( 1. + det * nu * (kk[np.newaxis, :]** 4  + ll[:, np.newaxis]** 4 ) )

    return mult * (ovar / det + rhs)

def lf(oovar, rhs, det,ll,kk,nu):
    """Leap frog timestepping: q^t-2 / ( 1 + 2. * dt * nu * (k^4 + l^4 ) ) + RHS"""
    mult = 2. * det / ( 1. + 2. * det * nu * (kk[np.newaxis, :] ** 4 + ll[:, np.newaxis] ** 4) )
    return mult * (oovar / det / 2. + rhs)

def filt(var, ovar, nvar,g):
    """Leapfrog filtering"""
    return var + g * (ovar - 2. * var + nvar )


########################################################
#new functionality 
########################################################
def calcrhs(qc1,qc2,q_1,q_2,vorc_1,vorc_2,psi1,psi2,psic_1,psic_2,psi_R,N,N2,ll,kk,tau_d,tau_f,sponge,beta,opt):
    
    #temp initialize
    forc1 = np.zeros(( N2, N))
    forc2 = np.zeros(( N2, N))
    cforc1 = np.zeros( ( N2, N // 2 + 1 ) ).astype(complex)
    cforc2 = np.zeros( ( N2, N // 2 + 1 ) ).astype(complex)
    nl1 = np.zeros( ( N2, N // 2 + 1 ) ).astype(complex)
    nl2 = np.zeros( ( N2, N // 2 + 1 ) ).astype(complex)
    
    
    if opt > 1:
#         print('in 1')
        #NL terms -J(psi, qc)
        nl1[:, :] = -nlterm( kk, ll, psic_1[1, :, :], vorc_1[1, :, :]) - beta * 1.j * np.expand_dims(kk, 0) * psic_1[1, :, :]        
        nl2[:, :] = -nlterm( kk, ll, psic_2[1, :, :], vorc_2[1, :, :]) - beta * 1.j * np.expand_dims(kk, 0) * psic_2[1, :, :]
        
    if opt != 2:
        #linear terms 
        #relax interface 
#         print('in 2')
        forc1[:, :] = (psi1[1] - psi2[1] - psi_R[:, np.newaxis]) / tau_d
        forc2[:, :] = -(psi1[1] - psi2[1] - psi_R[:, np.newaxis]) / tau_d
        
        #Sponge
        forc1[:, :] -= sponge[:, np.newaxis] * (q_1[1] - np.mean( q_1[1], axis = 1)[:, np.newaxis] )
        forc2[:, :] -= sponge[:, np.newaxis] * (q_2[1] - np.mean( q_2[1], axis = 1)[:, np.newaxis] )

        
        #Convert to spectral space + add friction
        cforc1 = fft.rfft2( forc1  , workers =nworker)
        cforc2 = fft.rfft2( forc2  , workers =nworker) + ( np.expand_dims(kk, 0) ** 2  + np.expand_dims(ll, 1) ** 2 ) * psic_2[1] / tau_f
    
    rhs1 = nl1[:] + cforc1[:]
    rhs2 = nl2[:] + cforc2[:]
    
    return rhs1,rhs2,cforc1,cforc2



```
